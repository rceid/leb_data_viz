```python
import geopandas as gpd
import pandas as pd
import zipfile
import io
import glob
import requests
import os
from shapely.geometry import Point
```

source: http://www.diva-gis.org/gdata


```python
URL = "http://biogeo.ucdavis.edu/data/diva/adm/LBN_adm.zip"
SHAPE_DIR = './shapes'
```

```python
if not os.path.exists(SHAPE_DIR):
    os.mkdir(SHAPE_DIR)
r = requests.get(URL)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall(SHAPE_DIR)
#[governate] = [f for f in z.namelist() if f == "LBN_adm1.shp"]
#[country] = [f for f in z.namelist() if f == "LBN_adm0.shp"]
[districts] = [f for f in z.namelist() if f == "LBN_adm2.shp"]
district_df = gpd.read_file(os.path.join(SHAPE_DIR, districts))
leb_df = pd.read_csv('./lebanon_food_prices_cleaned.csv')
#leb_df = leb_df.loc[leb_df['date'].apply(lambda date_: int(date_[:4]) > 2018)]
crs = {'init': 'epsg:4326'}
geometry = [Point(xy) for xy in zip(leb_df['long'].astype(float), leb_df['lat'].astype(float))]
leb_df = gpd.GeoDataFrame(leb_df,crs=crs, geometry=geometry)
leb_districts = gpd.sjoin(leb_df, district_df, how="inner", op='intersects')

[os.remove(f) for f in glob.glob(os.path.join(SHAPE_DIR, "*"))]
```
